```python
import cv2
import numpy as np
import os
from random import shuffle
from matplotlib import pyplot as plt
from tqdm import tqdm
import time as t

TRAIN_DIR = '/home/vlad/projects/deep-learning-nano/03-kaggle-dogs-vs-cats/train'
TEST_DIR = '/home/vlad/projects/deep-learning-nano/03-kaggle-dogs-vs-cats/test'

# image hiperparams
IMG_SIZE = 50

PREPARED_TRAIN_DATA_PATH='/home/vlad/projects/deep-learning-nano/03-kaggle-dogs-vs-cats/train_data.{}-beta.npy'.format(IMG_SIZE)
PREPARED_TEST_DATA_PATH='/home/vlad/projects/deep-learning-nano/03-kaggle-dogs-vs-cats/test_data.{}-beta.npy'.format(IMG_SIZE)


NUM_CHANELS = 3 # how many values per pixel (R,G,B) = 3or 0-1 from black to white = 1.
```

This section sets up function to read and prepare data from the dataset only once and then have it available as raw input for the model.

####Labeling 
We have a labeling function which checks to see if the word 'cat' is found in the given string. If so, it returns [1, 0] - indicating 'cat 100%' and 'dog 0%'. This data will be used as the expected values from the prediction model.

```python
# we check the name of the image to decide type of image (dog or cat)
# dog.01png. cat.01.png
def label_img(img):
    word_label = img.split('.')[0]
    if word_label == 'cat': return [1, 0]
    else: return [0, 1]
```

```python
def get_image(img_path):
    input_image = cv2.imread(img_path)
    
    height, width = input_image.shape[:2]
    
    if ( width > height ):
        if ( width > IMG_SIZE ):
            input_image = cv2.resize(
                input_image,
                (IMG_SIZE, int(IMG_SIZE/(width * 1.0) * height))
            )
    else:
        if ( height > IMG_SIZE ):
            input_image = cv2.resize(
                input_image,
                (int(IMG_SIZE/(height*1.0) * width), IMG_SIZE)
        )
    
    final_height, final_width = input_image.shape[:2]
    
    if ( final_width != IMG_SIZE or final_height != IMG_SIZE ):
        blank_image = np.zeros((IMG_SIZE, IMG_SIZE, 3), np.uint8)
        left_padding = int((IMG_SIZE - final_width)/2)
        top_padding = int((IMG_SIZE - final_height)/2)
        
        # https://stackoverflow.com/questions/14063070/
        # overlay-a-smaller-image-on-a-larger-image-python-opencv
        blank_image [ 
            top_padding : top_padding + final_height,
            left_padding : left_padding + final_width
        ] = input_image
        
        input_image = blank_image
        #cv_rgb = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
        #plt.imshow(cv_rgb)
        #plt.show()
        
    return input_image
```

```python
def read_and_prepare_input_images():
    training_data = []
    list_of_files = os.listdir(TRAIN_DIR)
    
    t0 = t.time()
    
    for img in tqdm(list_of_files):
        label = label_img(img)
        path = os.path.join(TRAIN_DIR, img)
        img = get_image(path)
        
        training_data.append([
           np.array(img),
           np.array(label)
        ])
        
    te = t.time()
    
    delta = te - t0
    
    print 'Operation done in %0.5f seconds' % delta
        
    return training_data
```

```python
def get_prepared_input_data(): 
    
    if os.path.isfile(PREPARED_TRAIN_DATA_PATH):
        training_data = np.load(PREPARED_TRAIN_DATA_PATH)
        return training_data
    
    training_data = read_and_prepare_input_images()
    shuffle(training_data)
    np.save(PREPARED_TRAIN_DATA_PATH, training_data)
    
    return training_data

training_data = get_prepared_input_data() 
```
